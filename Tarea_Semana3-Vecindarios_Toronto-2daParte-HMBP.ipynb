```python
# Importamos el paquete de geopy para determinar las coordenadas de Toronto City
!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim
```

```python
# Buscamos las coordenadas para realizar el mapa
address = 'Toronto, Ontario'
geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('Las coordenadas de '+address,' son {}, {}.'.format(latitude,longitude))
```

```python
# Importamos el paquete Folium para realizar el mapa
!pip install folium
import folium
```

```python
# Graficamos el mapa de Toronto City
map_toronto = folium.Map(location=[latitude, longitude], zoom_start = 10)
map_toronto
```

```python
# Añadimos marcadores de los municipios y vecindarios
for lat,lng, borough, neighborhood in zip(df_toronto['Latitude'],df_toronto['Longitude'],df_toronto['Borough'],df_toronto['Neighbourhood']):
    label= 'Neighborhood {}, Borough {}'.format(neighborhood,borough)
    label= folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup= label,
        color= 'red',
        fill=True,
        fill_color='yellow',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)

map_toronto
```

```python
# Información adicional: Listamos los municipios de Toronto y contamos las repeticiones , que nos indican cuantos vecindarios tiene cada uno
df_toronto_borough = pd.value_counts(df_toronto['Borough'])
df_toronto_borough = pd.DataFrame(df_toronto_borough).reset_index()
df_toronto_borough = df_toronto_borough.rename(columns={'index':'Borough','Borough':'Number of Neighborhood'})
df_toronto_borough
```
